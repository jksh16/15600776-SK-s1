```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = 'wine_quality.csv'  # Update with correct path
wine_data = pd.read_csv(file_path)

# Specify the features to be weighted
weighted_features = ['alcohol', 'citric.acid', 'free.sulfur.dioxide']
other_features = [col for col in wine_data.columns if col not in weighted_features + ['id', 'quality']]

# Create a new DataFrame with weighted features
wine_data_weighted = wine_data.copy()

# Apply weights to the selected features
wine_data_weighted[weighted_features] = wine_data_weighted[weighted_features] * 2

# Splitting the dataset
X_weighted = wine_data_weighted.drop(columns=['id', 'quality'])
y = wine_data_weighted['quality']
X_train, X_test, y_train, y_test = train_test_split(X_weighted, y, test_size=0.2, random_state=42)

# Scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Creating and fitting the Random Forest Classifier model
forest_weighted = RandomForestClassifier(random_state=42)
forest_weighted.fit(X_train_scaled, y_train)

# Making predictions with the weighted model
y_pred_weighted_rf = forest_weighted.predict(X_test_scaled)

# Calculating Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared (R2) for the model
mse_weighted_rf = mean_squared_error(y_test, y_pred_weighted_rf)
rmse_weighted_rf = mse_weighted_rf ** 0.5
r2_weighted_rf = r2_score(y_test, y_pred_weighted_rf)

# Printing the evaluation metrics
print(f"Mean Squared Error (MSE): {mse_weighted_rf:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_weighted_rf:.2f}")
print(f"R-squared (R2): {r2_weighted_rf:.2f}")
```
